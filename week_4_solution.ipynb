```r
library(tidyverse)
library(assertthat)
```

```r
#Question 1

sd_x= function(x){
    sqrt(sum(((x-mean(x))^2))/(length(x)-1))
}

x= c(1,2,3,4,5,6,7,8)
sd_x(x)
assert_that(sd_x(x)==sd(x))

y= c(20,30,40,50,60,70,80)
sd_x(y)
assert_that(sd_x(y)==sd(y))

z= c(1,5,10,15,20,25,30)
sd_x(z)
assert_that(sd_x(z)==sd(z))

```

```r
#Question 2
coral_tbl=as_tibble(read.csv(file = 'Kahekili_coral.csv'))

coral_pop = c("Montipora capitata",
"Porites compressa",
"Porites lobata")

coral_health_tbl = tibble(coral= sample(coral_pop, 100, replace=TRUE),
        health_status=sample(c(TRUE,FALSE), 100, replace=TRUE, prob=c(0.2,0.8)))

coral_health_tbl
```

```r
#Question 3

return_cover= function(x){
    if(x==TRUE) {
        rnorm(1,9,2)
    } else if (x==FALSE) {
        rnorm(1,1,3) 
    } else {
        stop("error")
    }
}
return_cover(TRUE)
return_cover(FALSE)
```

```r
coral_cover = mapply(return_cover, coral_health_tbl$health_status)

coral_health_tbl_2=add_column(coral_health_tbl, coral_cover)
coral_z=tibble(coral_health_tbl, coral_cover)

coral_health_tbl_2
coral_z
```

```r
#Question 4

binary_health_return= function(x){
    if(x==TRUE) {
        sum(1)
    } else if (x==FALSE) {
        sum(0) 
    } else {
        stop("error")
    }
}

binary_health_status = mapply(binary_health_return, coral_health_tbl$health_status)

coral_health_tbl_3=add_column(coral_health_tbl_2, binary_health_status)
coral_health_tbl_3
```
